```python
from sklearn.feature_selection import SelectKBest, f_regression
from sklearn.linear_model import LinearRegression
import cufflinks as cf
import pandas as pd
import numpy as np

cf.go_offline()
```

```python
df = pd.DataFrame(columns=[])
```

```python
df[0] = np.random.randint(200, size=1000)
df[1] = np.random.randint(200, size=1000)
```

```python
df["Xr"] = df[0] - df[0].mean()
df["Xr2"] = (df[0] - df[0].mean())**2
```

```python
df["yr"] = df[1] - df[1].mean()
```

```python
b_1 = ((df["Xr"] * df["yr"]).sum())/df["Xr2"].sum()
```

```python
b_1
```

```python
b_0 = df[1].mean() - b_1*df[0].mean()
```

```python
df["y_hat"] = b_0 + b_1*df[0]
```

```python
df["ri_2"] = (df[1] - df["y_hat"])**2
```

```python
rss = df["ri_2"].sum()
```

```python
tss = ((df[1] - df[1].mean())**2).sum()
```

```python
1-(rss/tss)
```

```python
df = pd.read_csv("/home/oscar/Desktop/Oscar/OnlineNewsPopularity.csv")
```

```python
df.columns = [x.strip() for x in df.columns]
```

```python
ls_var = ['n_tokens_title', 'n_tokens_content', 'num_imgs', 'num_videos', 
'data_channel_is_lifestyle', 'data_channel_is_entertainment', 
'data_channel_is_bus', 'data_channel_is_socmed', 
'data_channel_is_tech', 'data_channel_is_world', ]
```

```python
kb = SelectKBest(k=5, score_func=f_regression)
```

```python
kb.fit(X = df[ls_var], y = df["shares"])
```

```python
ls_best = [x for x, y in zip(ls_var, kb.get_support()) if y]
```

```python
ls_best
```

```python
lr = LinearRegression()
```

```python
lr.fit(X = df[ls_best], y = df["shares"])
```

```python
lr.score(X = df[ls_best], y = df["shares"])
```

```python
df["y_hat"] = lr.predict(X = df[ls_best])
```

```python
df[["shares", "y_hat"]]
```

```python
lr.intercept_
```

```python
[(x, y) for x, y in zip(ls_best, lr.coef_)]
```

```python
df["tgt"] = (df["shares"] > df["shares"].mean()).astype(int)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression(n_jobs=-1)
```
